## Local Inference on GPU
Model page: https://huggingface.co/suno/bark

⚠️ If the generated code snippets do not work, please open an issue on either the [model repo](https://huggingface.co/suno/bark)
			and/or on [huggingface.js](https://github.com/huggingface/huggingface.js/blob/main/packages/tasks/src/model-libraries-snippets.ts) 🙏

```python
# Use a pipeline as a high-level helper
from transformers import pipeline

pipe = pipeline("text-to-speech", model="suno/bark")
```

```python
# Load model directly
from transformers import AutoProcessor, AutoModelForTextToWaveform

processor = AutoProcessor.from_pretrained("suno/bark")
model = AutoModelForTextToWaveform.from_pretrained("suno/bark")
```

```python
from IPython.display import Audio

# Sample text to convert to speech
text = "नमस्ते, मेरा नाम बार्क है।"

# Generate speech
speech = pipe(text, forward_params={"do_sample": True})

# Save the audio to a file
output_filename = "bark_hindi_output.wav"
with open(output_filename, "wb") as f:
    f.write(speech["audio"][0].tobytes())

print(f"Audio saved to {output_filename}")

# Play the audio in Colab
Audio(speech["audio"], rate=speech["sampling_rate"])
```
